```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

```

```python
X_test.head()
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python


# Step 2: Convert encoded labels to one-hot-encoding
y
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

## Create Deep Learning MOdel

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create a Neural Network model here
model= Sequential()
model.add(Dense(units=100, activation='relu', input_dim= 53))
model.add(Dense(units=100, activation='relu'))  # Deep learning. -Hidden Layer
model.add(Dense(units=2, activation='softmax'))  # Number of -y- categories 
```

```python
# Compile and fit the model


# Print the 
model.summary()
```

```python
# Compile & Optimize the model

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=5,   # Number of iterations
    shuffle=True,
    verbose=2
)
```

## Quantify our Trained Model

```python
model_loss, model_accuracy = model.evaluate(
                    X_test_scaled,
                    y_test_categorical,
                    verbose=2)

print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Make Predictions


```python
# Use the first 5 test data values to make a prediction and compare it to the actual Labels

encoded_predictions= model.predict_classes(X_test_scaled[:10])
predicted_labels = label_encoder.inverse_transform(encoded_predictions)

print(encoded_predictions)
print(predicted_labels)
print(f"Actual Labels: {list(y_test[:6])}")  # Fisrt 5 elements predicted
```
