```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

```

```python
# Storing file in a path
hr = "Resources/WA_Fn-UseC_-HR-Employee-Attrition.csv"

# Read our Data file with the pandas library
hr_df = pd.read_csv(hr, encoding="ISO-8859-1")
hr_df
hr_df = hr_df.replace({"Yes": 1, "No": 0})
```

```python

import matplotlib.pyplot as plt

plt.matshow(hr_df.corr())
plt.show()
```

```python
f = plt.figure(figsize=(19, 15))
plt.matshow(hr_df.corr(), fignum=f.number)
plt.xticks(range(hr_df.select_dtypes(['number']).shape[1]), hr_df.select_dtypes(['number']).columns, fontsize=14, rotation=90)
plt.yticks(range(hr_df.select_dtypes(['number']).shape[1]), hr_df.select_dtypes(['number']).columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=13)
plt.title('Correlation Matrix', fontsize=15);
```
